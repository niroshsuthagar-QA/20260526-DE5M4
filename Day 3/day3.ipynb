```python
import pandas as pd
from sklearn.datasets import fetch_openml
```

```python
titanic = fetch_openml(name="titanic", version=1, as_frame=True)
data = pd.DataFrame(titanic.data)
data.head()
```

```python
# Q1
data.isnull().sum()

# Filling NA values
data["age"].fillna(data["age"].mean(), inplace=True)
data["cabin"].fillna("UNKNOWN", inplace=True)

# Dropping some of the NA values
data.dropna(subset=["embarked"], inplace=True)
```

```python
#Q2 Duplicate check
print(data.duplicated().sum())
data.drop_duplicates(inplace=True)
```

```python
# Q3 Finding inconsistencies 
data.sex.unique()
```

```python
# Q4 Finding outliers 

def outlierCheck(df, colName):
    q1  = df[colName].quantile(0.25)
    q3 = df[colName].quantile(0.75)

    IQR = q3 - q1

    # Outliers filtering
    outliers = df.loc[(df[colName] < (q1 - 1.5*IQR)) | (df[colName] > (q3 + 1.5*IQR))]

    return outliers
```

```python
outliers = outlierCheck(data, colName="age")
outliers.head(n=10)
```

```python
# Q5 Creating a new col 
data["family_size"] = data["sibsp"] + data["parch"]  + 1
data.head()
```

```python
# Q6 Encoding - Label Encoder
from sklearn.preprocessing import LabelEncoder
data.embarked.unique()

label_encoder = LabelEncoder()

data["embarked"] = label_encoder.fit_transform(data["embarked"])

data.tail()
```
